```python
import json
import networkx as nx

from py2cytoscape import util as cy 

!pip show py2cytoscape
```

```python
g = nx.Graph()
g.add_edge(1, 2, interaction='itr1',  score=0.1)
cyjs = cy.from_networkx(g)

print(json.dumps(cyjs, indent=4))
```

```python
# Test 2
g2 = nx.path_graph(3)
eb = nx.edge_betweenness(g2, normalized=False)
nx.set_edge_attributes(g2, 'eb', eb)
cyjs2 = cy.from_networkx(g2)

print(json.dumps(cyjs2, indent=4))
```
